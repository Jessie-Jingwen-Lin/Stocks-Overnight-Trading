```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('temp_stock_price.pkl', 'rb') as f:
    price_df = pickle.load(f)
```

```python
price_df
```

```python
recent_price_df = price_df.iloc[-20:]
```

```python
price_df
```

```python
morn_nulls = price_df['10:30'].isnull().mean(axis=0)
after_nulls = price_df['14:30'].isnull().mean(axis=0)

recent_morn_nulls = recent_price_df['10:30'].isnull().mean(axis=0)
recent_after_nulls = recent_price_df['14:30'].isnull().mean(axis=0)
```

```python
price_df['10:30']['AAPL']
```

```python
df = pd.DataFrame({'morning': morn_nulls, 'afternoon': after_nulls})
recent_df = pd.DataFrame({'morning': recent_morn_nulls, 'afternoon': recent_after_nulls})
```

```python
df
```

```python
df.mean()
```

```python
df.plot.scatter(x='morning', y='afternoon', marker='.')
```

```python
recent_df.plot.scatter(x='morning', y='afternoon', marker='.')
```

```python
df['avg'] = (df.morning + df.afternoon)/2
recent_df['avg'] = (recent_df.morning + recent_df.afternoon)/2
```

```python
df.avg.plot.hist()
```

```python
sns.scatterplot(data=df, x="morning", y="afternoon")
```

```python
plt.scatter(df.avg, recent_df.avg, marker='.')
```

```python
recent_df.avg.plot.hist()
```

```python
bad_tickers_nulls = recent_df[recent_df.avg >= 0.5]
```

```python
bad_tickers_nulls_list = list(bad_tickers_nulls.index)
```

```python
with open('temp_bad_tickers_nulls.pkl', 'wb') as f:
        pickle.dump(bad_tickers_nulls_list, f)
```

```python
with open('temp_all_tickers.pkl', 'wb') as f:
        pickle.dump(list(recent_df.index), f)
```
